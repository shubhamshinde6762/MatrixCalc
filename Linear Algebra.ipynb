```python
#@title $\color{yellow}{Input}$ { display-mode: "form" }
#@markdown ######$Enter \ No \ of \ \color{lime}{Unknowns}:$
n = 5 #@param {type:"integer"}
n+=1
#@markdown ######$Enter \ No \ of \ \color{lime}{Equations}:$
m = 3 #@param {type:"integer"}
#@markdown ######$Enter \ The \ Coefficients \ of  \ \color{lime}{Equations} \ like \ \color{cyan}{\ a_1,a_2,a_3...,b_1}:$

import numpy as np
import sympy as sp
from IPython.display import display, Latex, Math

#Function Defined For Input A Matrix
def suffix(i):
  if i+1==1:
    return 'st '
  elif i+1==2:
    return 'nd '
  elif i+1==3:
    return 'rd '
  else:
    return 'th'
def input_matrix(m,n):
  matrix=[]
  for i in range(m):
    int_list=[]
    input_str = input("Enter {}{} Row Of Matrix: ".format(i+1,suffix(i)))
    int_list = [sp.nsimplify(num) for num in input_str.split(",")[:n]]
    matrix.append(int_list)
  return matrix

#Function Defined For Converting a Matrix into Latex
def Latex_Convert(matrix):
  A = np.array(matrix)
  matrix_str = r'\begin{bmatrix}' + '\n'
  for row in A:
      matrix_str += ' & '.join([sp.latex(sp.simplify(elem)) for elem in row])
      matrix_str += r'\\' + '\n'
  matrix_str += r'\end{bmatrix}'
  return matrix_str

#To Convert A String Into Latex
def Latex_String(s):
  txt_1=str(s)
  txt_1=txt_1.replace(' ','\hspace{0.12cm}')
  txt_1='{'+txt_1+'}'
  return txt_1

#To InterChange Rows Of Matrix
def RowInterchange(matrix,i,j):
  txt_2="R_{} \Leftrightarrow R_{}".format(i,j)
  display(Latex('{'+txt_2+'}'))
  matrix[i-1], matrix[j-1] = matrix[j-1], matrix[i-1]
  return matrix

#Matrix Manupulation
def Reduce_Row(matrix,m,n):
  Rank=0
  Mn_Array=min(m,n)
  Count_Row_Interchanged=0
  List_Multiplied_Factor=[]
  for d in range (0,Mn_Array):
    Diagonal = matrix[d][d]
    for i in range(d+1,m):
      if (matrix[d][d]!=0):
        if (matrix[i][d] != 0):
          ele_mul_to_Dia =matrix[i][d]
          txt="{}".format(sp.latex(matrix[i][d]))
          if(matrix[i][d]<0):
            txt='('+txt+')'
          arr = np.array(matrix)
          arr[i] = Diagonal*arr[i] - matrix[i][d]*arr[d]
          matrix_str = r'\begin{bmatrix}' + '\n'
          for row in arr:
              matrix_str += ' & '.join(sp.latex(sp.simplify(elem)) for elem in row)
              matrix_str += r'\\' + '\n'
          matrix_str += r'\end{bmatrix}'
          matrix = arr.tolist()
          Lat_A = matrix_str

          List_Multiplied_Factor.append(Diagonal)

          txt2="{}".format(sp.latex(Diagonal))
          txt3="R_{}".format(i+1)
          txt4=r"R_{} \Rightarrow ".format(i+1)
          txt5=" R_{}".format(sp.Rational(d+1))
          if(Diagonal==1):
              Main_txt=txt4+txt3+"-"+txt+txt5
          elif(ele_mul_to_Dia==1):
              Main_txt=txt4+txt2+txt3+"-"+txt5
          else:
              Main_txt=txt4+txt2+txt3+"-"+txt+txt5
          print('\n')
          display(Latex(Main_txt))
          print('\n')
          display(Latex("\hspace{1.5cm}{Aug(A|B)=}"+Lat_A))
        else:
          None
      else:
        Row_Remain=0
        row=d
        garbage=matrix.copy()
        while(Row_Remain<(m-d-1) and garbage[d][d]==0):
          row+=1
          garbage[d],garbage[row]=garbage[row],garbage[d]
          Row_Remain+=1
          if(garbage[d][d]!=0):
            print('\n')
            matrix=RowInterchange(matrix,d+1,row+1)
            Lat_A1 = Latex_Convert(matrix)
            print('\n')
            display(Latex("\hspace{1.5cm}{Aug(A|B)=}"+Lat_A1))
            Diagonal=matrix[d][d]
            Count_Row_Interchanged+=1
            i-=1
            break

  pivote=[]
  piv=0
  for d in range(0,min(m,n)-1):
    if (matrix[d][d]==0):
      for col in range(d+1,n):
        a=matrix[d][col]
        row=d
        b=0
        while(m-row>0):
          if(matrix[row][col]!=0):
            piv=col
            if(piv!=col):
              pivote.append(piv)
            if(d!=row and a==0):
              matrix=RowInterchange(matrix,row+1,d+1)
              Lat_A1 = Latex_Convert(matrix)
              print('\n')
              display(Latex("\hspace{1.5cm}{Aug(A|B)=}"+Lat_A1))
              Count_Row_Interchanged+=1
              a=matrix[d][col]
              b=1
              continue
            elif(d!=row and a!=0):
              arr = np.array(matrix)


              txt="{}".formatsp.latex((matrix[row][col]))

              if(matrix[row][col]<0):
                txt='('+txt+')'

              List_Multiplied_Factor.append(a)

              arr[row] = a*arr[row] - matrix[row][col]*arr[d]
              matrix = arr.tolist()


              txt2="{}".format(sp.latex(a))
              txt3="R_{}".format(row+1)
              txt4=r"R_{} \Rightarrow ".format(row+1)
              txt5=" R_{}".format(d+1)

              if(a==1):
                Main_txt=txt4+txt3+"-"+txt+txt5
              elif(matrix[row][col]==1):
                Main_txt=txt4+txt2+txt3+"-"+txt5
              else:
                Main_txt=txt4+txt2+txt3+"-"+txt+txt5
              print('\n')
              display(Latex(Main_txt))
              print('\n')
              display(Latex("\hspace{1.5cm}{Aug(A|B)=}"+Latex_Convert(matrix)))
            b=1

          else:
            pass
          row+=1
        if(b==1):
          break
  return matrix,List_Multiplied_Factor,Count_Row_Interchanged,Rank,pivote



#To Get Echelon Form
def RowEch(matrix,m,n):
  Lat_A = Latex_Convert(matrix)
  txt_1=Latex_String('Let,')
  display(Latex(txt_1))
  print('\n')
  display(Latex("\hspace{1.5cm}{AX=B}"))
  print('\n')
  display(Latex("\hspace{1.5cm}{Aug(A|B)=}"+Lat_A))
  matrix,List_Multiplied_Factor,Count_Row_Interchanged,Rank,pivote=Reduce_Row(Mat_A,m,n)
  print('\n')

  display(Latex(Latex_String('   Thus, Echelon Form is,')))
  print('\n')
  Lat_A = Latex_Convert(matrix)
  display(Latex("\hspace{1.5cm}{Aug(A|B)=}"+Lat_A))
  return matrix,List_Multiplied_Factor,Count_Row_Interchanged,Rank,pivote

Mat_A=input_matrix(m,n)
```

#$\color{yellow} {Row\ Echelon\ Form \ of \ Augmented \ Matrix \ \rightarrow }$

$ \textbf{Step 1:} \ Start \ with \ the \ original \ matrix.$

$\textbf{Step 2:} \ Locate \  the \  left \ most  \ column \  that  \ has \  a \ non \ zero  \
entry  \ in \  any  \ {row.}  \ This  \ will  \ be  \ the \  \  pivot  \ {column.}$

$\textbf{Step 3:} \ {(\color{lime}{Optional})} \ Make \ the \ pivot \ entry \ equal \ to \ {1.} \ If \  it's \ not \ already \ {1,} \ you \ can \ divide \ the \ entire \ row \ by \ the \ pivot \ entry \ to \ achieve \ {this.}$

$\textbf{Step 4:} \ Use \ row \ operations \ to \ create \ zeros \ below \ the \ pivot \ {entry.} \ To \ do \ {this,} \ subtract \ multiples \ of \ the \ pivot \ row \ from \ the \ rows \ below \ {it.}$

$\textbf{Step 5:} \ Move \ to \ the \ next \ column \ to \ the \ right \ and \ repeat \ steps \ {2-4} \ until \ you \ have \ processed \ all \ columns \ or \ you \ have \ reached \ the \\ \ \ \ \ \ \ \ \ \ \ \ \ \ \ bottom \ of \ the \ {matrix.}$

$\textbf{Step 6:} \ If \ there \ are\ any \ rows \ with \ all \ {zeros,} move \ them \ to \ the \ bottom \ of \ the \ {matrix.}$

$\textbf{Step 7:} \ Repeat \ steps \ {2-6} \ on \ the \ submatrix \ formed \ by \ removing \ the \ first \ column \ and \ the \ rows \ below \ the \ {pivot.}$

$\textbf{Step 8:} \ Continue \ this \ process \ until \ you \ have \ processed \ all \ columns \ or \ you \ have \ reached \ the \ {bottom-right} \ corner \ of \ the \ {matrix.}$
$\textbf{Step 9:} \ The \ resulting \ matrix \ will \ be \ in \ row \ echelon \ {form.}$
